```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

```python
df = pd.read_csv('concrete.csv',index_col=0)
```

```python
df.columns
```

```python
# create a new figure
fig = plt.figure()

ax = fig.add_subplot(111,projection='3d')
ax.set_xlabel('superplastic')
ax.set_ylabel('Water')
ax.set_zlabel('Strength')
# C stands for color.
# marker: set an open circle marker.(^)
ax.scatter(df.superplastic, df.water, df.strength, c='r', marker='o')
plt.suptitle('superplastic vs Water vs Strength')

# plt.show() not needed if you use '%matplotlib inline'
plt.show()
```
