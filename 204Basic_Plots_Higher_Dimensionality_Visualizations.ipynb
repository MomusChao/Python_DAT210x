# Andrew's Curves

Andrews plot, also known as Andrews curve. It help you visualize higher dimensionality, multi-varied data by plotting your data as a set of curves. The feature values for your samples actually act as the coefficients of the curve, so observations with similar characteristics tend to group closer to each other.

```python
from sklearn.datasets import load_iris
from pandas.tools.plotting import andrews_curves
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Look pretty...
matplotlib.style.use('ggplot')
# If the above line throws an error, use plt.style.use('ggplot') instead
```

```python
# Load up SKLearn's Iris Dataset into a Pandas Dataframe
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target_names'] = [data.target_names[i] for i in data.target]
```

```python
# Andrews Curves Start Here:
plt.figure()
andrews_curves(df, 'target_names')
plt.show()
```

One of the current weaknesses with the Pandas implementation is that every single observation is charted.
