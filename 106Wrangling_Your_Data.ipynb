# Module2 - Lab3

```python
# Import and alias Pandas
import pandas as pd
```

```python
# Load up the Servo.data dataset
df = pd.read_csv("servo.data", header=None)
df.columns = ['motor', 'screw', 'pgain', 'vgain', 'class']
```

```python
# slicing
vgain5 = df[df.vgain == 5]
vgain5.count()
```

```python
# Containing the value E for both motor and screw features
condition = (df.motor == 'E') & (df.screw == 'E')
motor_screw_E = df[condition]
motor_screw_E.count()
```

```python
# the mean vgain value of those samples that have a pgain feature value equal to 4
pgain4 = df[df.pgain == 4]
pgain4.vgain.mean()
```

# Module2 - Lab4

```python
import pandas as pd
```

```python
dfs = pd.read_html("http://www.espn.com/nhl/statistics/player/_/stat/points/sort/points/year/2015/seasontype/2")
```

```python
# rename the columns
df = dfs[0]
header = ['RK','PLAYER','TEAM','GP','G','A','PTS','PN','PIM','PTSG','SOG','PCT','GWG','PPG','PPA','SHG','SHA']
df = df[2:]
df.columns = header
```

```python
# Get rid of (drop) any erroneous rows that has at least 4 NANs in them
df = df.dropna(axis=0,thresh=3)
```

```python
# Get rid of the RK column
condition = (df.RK != 'RK')
df = df[condition]
df = df.drop('RK',axis=1)
```

```python
# Ensure there are no nan "holes" in your index
df.reset_index(drop=True)
```

```python
# Check the dtypes of all columns, and ensure those that should be numeric are numeric
df['G'] = pd.to_numeric(df['G'])
df['A'] = pd.to_numeric(df['A'])
df['PTS'] = pd.to_numeric(df['PTS'])
df['PN'] = pd.to_numeric(df['PN'])
df['PIM'] = pd.to_numeric(df['PIM'])
df['PTSG'] = pd.to_numeric(df['PTSG'])
df['SOG'] = pd.to_numeric(df['SOG'])
df['PCT'] = pd.to_numeric(df['PCT'])
df['GWG'] = pd.to_numeric(df['GWG'])
df['PPG'] = pd.to_numeric(df['PPG'])
df['PPA'] = pd.to_numeric(df['PPA'])
df['SHG'] = pd.to_numeric(df['SHG'])
df['SHA'] = pd.to_numeric(df['SHA'])
```
